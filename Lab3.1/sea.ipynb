```python
import pygame
from pygame.draw import *
import numpy as np

# Почти все функции принимают xl, yd, xr, yu. Это координаты левой границы избражения, 
# нижней, правой и верхней соответственно.

def cloud(xl, yd, xr, yu, n, z_x_intersection = 0.5, z_y_intersection = 0.5):  
    #облака рисуют не там где предпологалось, ну и ладно.
    s = pygame.Surface((700,400), pygame.SRCALPHA)
    x_inters = int(z_x_intersection * (xr - xl)/n)
    R = (xr - xl)//(2*n)
    y_inters = int(((yd - yu)/2)*z_y_intersection)
    y_R = int(((yd - yu)/2)*(1 + z_y_intersection))
    for i in range(n):
        ellipse(s, (255, 255, 255, 128), [xl + 2*i*R - x_inters*i, yd - 2*y_R, 2*R, y_R])
        ellipse(s, BLACK, [xl + 2*i*R - x_inters*i, yd - 2*y_R, 2*R, y_R], 1)

    for k in range(n-1):
        ellipse(s, (255, 255, 255, 128), [xl + 2*k*R + int(0.5*R) - x_inters*k, yd - 2*y_R - y_R + y_inters, 2*R, y_R])
        ellipse(s, BLACK, [xl + 2*k*R + int(0.5*R) - x_inters*k, yd - 2*y_R - y_R + y_inters, 2*R, y_R], 1)
    screen.blit(s, (0,0))


def circle_4(x, y, R, C):
    masiv = []
    for i in range(20):
        angle = (np.pi/39)*i
        masiv.append((x - R*np.sin(angle), y + R*np.cos(angle)))
    masiv.append((x,y))
    polygon(screen, C, masiv)


def treugolnik( x1, y1, x2, y2): 
    xp = x1 - x2
    yp = y2 - y1
    polygon(screen, RED, [(x1, y1), (x2, y2), (x1, y2)])
    for i in range(5):
        x = -xp/4*i + x1
        line(screen, BLACK, (x1, y1), (x, y2))
    


def ship(xl, yd, xr, nomer):
    xp = xr - xl
    yu = yd - xp
    yp = yd - yu
    x3 = xl + 1/2*xp    
    if nomer == 1:
        C = BROWN
        z_d = 1/5
        x5 = xr - 1/5*xp
        x4 = x3 - 1/50*xp
        xR = xr - 1/6*xp
    if nomer == 2:
        C = ORANGE
        z_d = 2/5
        x5 = xr - 3/5*xp
        x4 = x3 - 1/50*xp
        xR = xr - 4/6*xp

    z_xp = 1/4
    z_yp = 1/5
    y2 = yd - int(yp*z_yp)
    x_1 = xl + int(z_d*xp)
    x2 = x_1 + int((1 - z_xp - z_d)*xp)
    rect(screen, C, [xl + int(z_d*xp), yd - int(yp*z_yp), int((1 - z_xp - z_d)*xp), int(z_yp*yp)])
    polygon(screen, C, [(x2, yd), (x2, y2), (xr, y2)])
    circle_4(x_1, y2, 1/5*xp, C)
    x3 = xl + 1/2*xp
    #x4 = x3 + 1/50*xp
    y3 = 1/3*yp + yu
    rect(screen, BLACK, [x3, y3, x4-x3, y2 - y3])
    y4 = (y3 + y2)/2
    x6 = x4 +(x5-x4)/3
    polygon(screen, GREY, [(x4,y3), (x6, y4), (x5, y4)] )
    polygon(screen, BLACK, [(x4,y3), (x6, y4), (x5, y4)], 1 )
    polygon(screen, GREY, [(x4,y2), (x6, y4), (x5, y4)] )
    polygon(screen, BLACK, [(x4,y2), (x6, y4), (x5, y4)], 1 )

    yR = yd - 1/7*yp
    circle(screen, White, (int(xR),int(yR)), int(xp*1/20))
    circle(screen, BLACK, (int(xR),int(yR)), int(xp*1/20), int(xp*1/40))


def zont(xl, yd, xr, yu):
    xp = xr - xl
    yp = yd - yu
    yc = yu + 1/4*yp
    xm = xp/2 + xl
    x_1 = xm - 1/20*xp
    x2 = xm + 1/20*xp
    rect(screen, RED,[x_1, yc, x2 - x_1, yu - yc])
    rect(screen, ORANGE, [x_1, yd, x2 - x_1, yc - yd])
    treugolnik(x_1, yu, xl, yc)
    treugolnik(x2, yu, xr, yc)
    

def Sun(x,y,R):
    masiv = []
    for i in range(50):
        alpha = 2*np.pi/49*i
        masiv.append(((x + R*np.sin(alpha)), (y + R*np.cos(alpha))))
        masiv.append(((x + 1.1*R*np.sin(alpha + 2*np.pi/98)), (y + 1.2*R*np.cos(alpha + 2*np.pi/98))))
    polygon(screen, YEL2, masiv)


def sinus(y_c):
    masiv = []
    for i in range(100):
        x = 700/99*i
        y = y_c + 10*np.sin(x)
        masiv.append((x,y))
    masiv.append((700, 400))
    masiv.append((0, 400))
    polygon(screen, YEL1, masiv)


def volna (color, y_c, H, f):
    masiv = []
    for i in range(100):
        x = 700/99*i
        y = y_c + H*np.sin(x + f)
        masiv.append((x,y))
    masiv.append((700, y + 50))
    masiv.append((0, y + 50))    
    polygon(screen, color, masiv)


def risunok():
    rect(screen,(117, 187, 253), [(0,0),(700,400)])
    volna((200, 150, 160), 195, 5, 1 - t)
    volna((200, 50, 60), 205, 5, np.pi - t/2)
    volna((100, 10, 160), 220, 5, 3 - t/3)
    volna((100, 10, 60), 240, 5, 5 - t)
    volna((142, 70, 60), 260, 5, 2 - 2*t)
    sinus(285)
    Sun(600, 100, 60)
    cloud(0 + 2*t, 100, 200 + 2*t, 50, 4)
    cloud(300 + 3*t, 130, 450 + 3*t, 70, 3)
    cloud(200 + 10*t, 200, 350 + 10*t, 150, 2)
    cloud(600 + t, 200, 700 + t, 150, 2)
    ship(50 + 7*t, 230, 150 + 7*t, 1)
    ship(450 - 20*t, 250, 650 - 20*t, 2)
    if t > 10:
        ship(-210 + 10*(t-10), 255, 0 + 10*(t-10), 1)
    zont(100, 350, 150, 250)
    zont(300, 370, 400, 190)

    
pygame.init()

t=0
FPS = 30
screen = pygame.display.set_mode((700, 400))
RED = (255, 0, 0)
BLACK = (0, 0, 0)
YEL1 = (210, 210, 0)
YEL2 = (255, 255, 0)
White = (255, 255, 255)
BROWN = (101, 67, 33)
GREY = (200, 200, 200)
ORANGE = (255, 165, 0)
BLUE = (0,0,255)

pygame.display.update()
clock = pygame.time.Clock()
finished = False

while not finished:
    clock.tick(FPS)
    t+=1/30
    if t > 50:

        t = 0
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            finished = True
    risunok()
    pygame.display.flip()


pygame.quit()
```
