```python
import pygame
from pygame.draw import *
from random import randint
from random import random
import angry_smail12
pygame.init()

FPS = 30
screen = pygame.display.set_mode((1200, 900))

RED = (255, 0, 0)
BLUE = (0, 0, 255)
YELLOW = (255, 255, 0)
GREEN = (0, 255, 0)
MAGENTA = (255, 0, 255)
CYAN = (0, 255, 255)
BLACK = (0, 0, 0)
COLORS = [RED, BLUE, YELLOW, GREEN, MAGENTA, CYAN]
xb = 50
yb = 200
width_b = 100
height_b = 50
x_name = 50
width_name = 400
y_name = 0
height_name = 50
count_name = True
name = ''
count_chit = False


def score(x, y):
    global level
    width = 200
    height = 50
    rect(screen, (0, 200, 0), (x, y, width, height))
    font = pygame.font.SysFont('serif', 48)
    text = font.render("level = " + str(level), 1, (0, 0, 0))
    screen.blit(text, (x, y))

def chit(x, y):
    global level, xb, yb, width_b, height_b, points, chit
    if xb < x < xb + width_b and yb < y < yb + height_b:
        points = k + number_smails
        count_chit = True

def botton():
    global xb, yb, width_b, height_b
    rect(screen, (200, 200, 0), (xb, yb, width_b, height_b))
    font = pygame.font.SysFont('serif', 48)
    text = font.render("cheat", 1, (0, 0, 0))
    screen.blit(text, (xb, yb))
    

    
def create_name(x, y):
    global x_name, width_name, y_name, height_name, count_name, name
    if x_name < x < x_name + width_name and y_name < y < y_name + height_name and count_name:
            fin = False
            while not fin:
                for event in pygame.event.get():
                    if event.type == pygame.KEYDOWN:
                        if event.key == pygame.K_RETURN:
                            fin = True
                        else:
                            name += event.unicode
            count_name = False
    


def name_window():
    global x_name, width_name, y_name, height_name
    rect(screen, (250, 250, 250), (x_name, y_name, width_name, height_name))
    font = pygame.font.SysFont('serif', 48)
    text = font.render("name: " + name, 1, (0, 0, 0))
    screen.blit(text, (x_name, y_name))
    
    
def bounce_off_the_border(i,  massiv_coordinat, massiv_scorostey):
    if massiv_coordinat[i][0] < 0 and massiv_scorostey[i][0] < 0:
        massiv_scorostey[i][0] *= -1
    if massiv_coordinat[i][0] > 1200 and massiv_scorostey[i][0] > 0  :
        massiv_scorostey[i][0] *= -1
    if massiv_coordinat[i][1] > 800 and massiv_scorostey[i][1] > 0 :
        massiv_scorostey[i][1] *= -1
    if massiv_coordinat[i][1] < 0 and massiv_scorostey[i][1] < 0:
        massiv_scorostey[i][1] *= -1
        
    if massiv_coordinat[i][0] < -15:
        massiv_coordinat[i][0] += 20
    if massiv_coordinat[i][0] > 1215:
        massiv_coordinat[i][0] -= 20
    if massiv_coordinat[i][1] > 815:
        massiv_coordinat[i][1] -= 20
    if massiv_coordinat[i][1] < -15:
        massiv_coordinat[i][1] += 20


def new_ball(massiv_scorostey, massiv_coordinat, angry = False):
    '''рисует новый шарик '''
    x_new = randint(100, 1100)
    y_new = randint(100, 800)
    if not(angry):
        r_new = randint(50, 100)
    else:
        r_new = 25*randint(2, 4)
        
    V_x = V*(0.5 - random()) 
    V_y = V*(0.5 - random())
    color = COLORS[randint(0, 5)]
    massiv_coordinat.append([x_new, y_new, r_new, color])
    massiv_scorostey.append([V_x, V_y])



def click(event,k , massiv_coordinat):
    global points, name
    pos = pygame.mouse.get_pos()
    x, y = pos[0], pos[1]
    for i in range(k):
        delta_x = x - massiv_coordinat[i][0]
        delta_y = y - massiv_coordinat[i][1]
        delta = (delta_x**2 + delta_y**2)**(1/2)
        if delta < massiv_coordinat[i][2]:
            massiv_coordinat[i] = [0, 0, 0, (0, 0, 0)]
            massiv_scorostey[i] = [0, 0]
            points += 1
    for i in range(k, k + number_smails):
        delta_x = x - massiv_coordinat[i][0]
        delta_y = y - massiv_coordinat[i][1]
        delta = (delta_x**2 + delta_y**2)**(1/2)
        if delta < massiv_coordinat[i][2]:
            massiv_coordinat[i] = [0, 0, 0, (0, 0, 0)]
            massiv_scorostey[i] = [0, 0]
            points += 1
    
    chit(x, y)
    create_name(x, y)
            


def ball(i, t, massiv_coordinat, massiv_scorostey):
    massiv_coordinat[i][0] = massiv_coordinat[i][0] + massiv_scorostey[i][0]
    massiv_coordinat[i][1] = massiv_coordinat[i][1] + massiv_scorostey[i][1]
    circle(screen, massiv_coordinat[i][3], (int(massiv_coordinat[i][0]),\
                                            int(massiv_coordinat[i][1])), massiv_coordinat[i][2])
    bounce_off_the_border(i, massiv_coordinat, massiv_scorostey)


def smail(i, t, massiv_coordinat, massiv_scorostey):
    global screen, a
    massiv_coordinat[i][0] = massiv_coordinat[i][0] + massiv_scorostey[i][0]
    massiv_coordinat[i][1] = massiv_coordinat[i][1] + massiv_scorostey[i][1]
    z = 100//(massiv_coordinat[i][2])
    angry_smail12.angry_smail(screen, int(massiv_coordinat[i][0]), int(massiv_coordinat[i][1]), z)
    bounce_off_the_border(i, massiv_coordinat, massiv_scorostey)
    
    pos = pygame.mouse.get_pos()
    x, y = pos
    delta_x = x - massiv_coordinat[i][0]
    delta_y = y - massiv_coordinat[i][1]
    delta = (delta_x**2 + delta_y**2)**(1/2)
    massiv_scorostey[i][0] -= a*delta_x/delta
    massiv_scorostey[i][1] -= a*delta_y/delta
    if (massiv_scorostey[i][0]**2 + massiv_scorostey[i][1]**2)**(1/2) > V:
        massiv_scorostey[i][0] += a*delta_x/delta
        massiv_scorostey[i][1] += a*delta_y/delta
    

def main(t, k, massiv_coordinat, massiv_scorostey):
    for i in range(k):
        ball(i, t, massiv_coordinat, massiv_scorostey)
    for i in range(k, k + number_smails):
        if not(massiv_coordinat[i][2] == 0):
            smail(i, t, massiv_coordinat, massiv_scorostey)
    score(50, 60)
    botton()
    name_window()


massiv_coordinat = []
massiv_scorostey = []
k = 3
number_smails = 2
t = 0
V = 0
a = 0.5
points = 0
level = 0

pygame.display.update()
clock = pygame.time.Clock()
finished = False

for m in range (k):
    new_ball(massiv_scorostey, massiv_coordinat)

for m in range(number_smails):
    new_ball(massiv_scorostey, massiv_coordinat, True)

while not finished:
    clock.tick(FPS)
    t += 1
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            finished = True
        elif event.type == pygame.MOUSEBUTTONDOWN:
            click(event, k, massiv_coordinat)

    main(t, k, massiv_coordinat, massiv_scorostey)
    pygame.display.update()
    screen.fill(BLACK)
    if points == k + number_smails:
        massiv_scorostey = []
        massiv_coordinat = []
        points = 0
        t = 0
        V += 10
        a += 0.5
        level += 1
        for m in range (k):
            new_ball(massiv_scorostey, massiv_coordinat)
        for m in range(number_smails):
            new_ball(massiv_scorostey, massiv_coordinat, angry = True)

pygame.quit()
if name == '':
    name = 'Аноним'
table = open('Таблица игроков.txt', 'a')
if not(count_chit):
    print(name, ':', level, file = table)
else:
    print(name, 'Читер', file = table)
```
