```python
import pygame
from pygame.draw import * 
def Leg(surface,color, x, y, width, hight):
    '''
    Рисует ногу.
    
    -----------
    Parametrs:
    
    surface : Surface
        Холст pygame
    
    x : int 
        Координата левой границы картинки.
        
    y : int
        Координата верхней границы картинки.
        
    width : int
        Ширина картинки.
        
    height : int
        Высота изображения.
        
    color : tuple(int, int, int) 
        цвет RGB
    
    ----------
    Возвращаемые значения: 
        None
    '''
    ellipse(surface, color, [x, y, width, hight])


def Body(surface, color, x, y, width, hight):
    '''
    Рисует тело.
    
    -----------
    Parametrs:
    
    surface : Surface
        Холст pygame
    
    x : int 
        Координата левой границы картинки.
        
    y : int
        Координата верхней границы картинки.
        
    width : int
        Ширина картинки.
        
    height : int
        Высота изображения.
        
    color : tuple(int, int, int) 
        цвет RGB
    
    ----------
    Возвращаемые значения: 
        None
    '''
    ellipse(surface, color, [x, y, width, hight])


def Ear(surface, color, x, y, width, hight):
    '''
    Рисует ушко.
    
    -----------
    Parametrs:
    
    surface : Surface
        Холст pygame
    
    x : int 
        Координата левой границы картинки.
        
    y : int
        Координата верхней границы картинки.
        
    width : int
        Ширина картинки.
        
    height : int
        Высота изображения.
        
    color : tuple(int, int, int) 
        цвет RGB
    
    ----------
    Возвращаемые значения: 
        None
    '''
    ellipse(surface, color, [x, y, width, hight])

def Eye(surface, x, y, width, hight):
    '''
    Рисует голубой глаз.
    
    -----------
    Parametrs:
    
    surface : Surface
        Холст pygame
    
    x : int 
        Координата левой границы картинки.
        
    y : int
        Координата верхней границы картинки.
        
    width : int
        Ширина картинки.
        
    height : int
        Высота изображения.
    
    ----------
    Возвращаемые значения: 
        None
    '''
    ellipse(surface, (117, 187, 253), [x, y, width, hight])
    ellipse(surface, (0, 0, 0), [x + width//3, y +hight//3, width//5, hight//5])


def Face(surface, color, x, y, width, hight):
    '''
    Рисует лицо.
    
    -----------
    Parametrs:
    
    surface : Surface
        Холст pygame
    
    x : int 
        Координата левой границы картинки.
        
    y : int
        Координата верхней границы картинки.
        
    width : int
        Ширина картинки.
        
    height : int
        Высота изображения.
        
    color : tuple(int, int, int) 
        цвет RGB
    
    ----------
    Возвращаемые значения: 
        None
    '''
    ellipse(surface, color, [x, y, width, hight])
    hight_E = hight//5
    width_E = width//5
    x_El = x + width_E
    x_Er = x + 3*width_E
    y_E = y + hight//2 - hight_E
    Eye(surface, x_El, y_E, width_E, hight_E)
    Eye(surface, x_Er, y_E, width_E, hight_E)
    rect(surface, (0, 0, 0), [x_El, y + hight//2, x_Er - x_El, hight//10])


def Head(surface, color, x, y, width, hight):
    '''
    Рисует голову.
    
    -----------
    Parametrs:
    
    surface : Surface
        Холст pygame
    
    x : int 
        Координата левой границы картинки.
        
    y : int
        Координата верхней границы картинки.
        
    width : int
        Ширина картинки.
        
    height : int
        Высота изображения.
        
    color : tuple(int, int, int) 
        цвет RGB
    
    ----------
    Возвращаемые значения: 
        None
    '''
    x_F = x
    hight_F = int(0.4*hight)
    y_F = y + hight - hight_F
    width_F = width
    Face(surface, color, x_F, y_F, width_F, hight_F)
    
    y_e = y
    width_e = width//4
    hight_e = int(1.2*(hight - hight_F))
    x_el = x + width//4 - width_e//2
    x_er = x + 3*width//4 - width_e//2
    Ear(surface, color, x_el, y_e, width_e, hight_e)
    Ear(surface, color, x_er, y_e, width_e, hight_e)
    
    
def Krolick(surface,color, x, y, width, hight):
    '''
    Рисует зайца.
    
    -----------
    Parametrs:
    
    surface : Surface
        Холст pygame
    
    x : int 
        Координата левой границы картинки.
        
    y : int
        Координата верхней границы картинки.
        
    width : int
        Ширина картинки.
        
    height : int
        Высота изображения.
        
    color : tuple(int, int, int) 
        цвет RGB
    
    ----------
    Возвращаемые значения: 
        None
    '''
    x_Ll = x
    hight_L = hight//20
    y_L = y + hight - hight_L
    x_Lr = x + 2*width//3
    width_L = width//3
    Leg(surface, color, x_Ll, y_L, width_L, hight_L)
    Leg(surface, color, x_Lr, y_L, width_L, hight_L)
    
    x_B = x + width_L//2
    width_B = 2*width//3
    hight_B = hight//2
    y_B = y + hight - hight_B
    Body(surface, color, x_B, y_B, width_B, hight_B)
    
    hight_H = hight - hight_B
    y_H = y
    width_H = width_B
    x_H = x + (width - width_B)//2
    Head(surface, color, x_H, y_H, width_H, hight_H)


FPS = 30
screen = pygame.display.set_mode((700, 400))
rect(screen,(0, 0, 0), [(0,0),(700,400)])
BLACK = (0, 0, 0)
White = (255, 255, 255)

Krolick(screen, White, 400, 30, 40, 200)
Krolick(screen, White, 10, 50, 150, 100)
Krolick(screen, White, 500, 150, 50, 150)
Krolick(screen, White, 300, 0, 100, 300)
Krolick(screen, White, 200, 0, 100, 400)


pygame.display.update()
clock = pygame.time.Clock()
finished = False

while not finished:
    clock.tick(FPS)
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            finished = True

pygame.quit()
```
